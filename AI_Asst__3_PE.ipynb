```
#After the req. changes
# fiilling the mean value where ever the 'NaN' appears in complete column 
weight_avg=round(df["weight"].mean(),2)
stand_by_time_avg=round(df["stand_by_time"].mean(),2)
no_of_reviews_in_gsmarena_in_week_avg=round(df["no_of_reviews_in_gsmarena_in_week"].mean(),2)
phone_height_avg=round(df["phone_height"].mean(),2)
talk_time_avg=round(df["talk_time"].mean(),2)
video_resolution_avg=round(df["video_resolution"].mean(),2)
primary_camera_resolution_avg=round(df["primary_camera_resolution"].mean(),2)

#  Replaceing the 'NaN' to column mean values
df["weight"].fillna(weight_avg, inplace=True)
df["stand_by_time"].fillna(stand_by_time_avg, inplace=True)
df["no_of_reviews_in_gsmarena_in_week"].fillna(no_of_reviews_in_gsmarena_in_week_avg, inplace=True)
df["phone_height"].fillna(phone_height_avg, inplace=True)
df["talk_time"].fillna(talk_time_avg, inplace=True)
df["video_resolution"].fillna(video_resolution_avg , inplace=True)
df["primary_camera_resolution"].fillna(primary_camera_resolution_avg, inplace=True)
```
